```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
rng = random.Generator(random.PCG64())
```

## Preparing a noisy sine wave

```python
x = np.arange(0, 3.5*np.pi, 0.1)
y = np.sin(x) + (rng.random(x.shape)-0.5)/2
y_og = np.sin(x)

plt.scatter(x,y)
plt.plot(x,y_og, color='red')
xticks = np.arange(0, 3.5*np.pi, np.pi/2)
xlabels = [f'{_} $\pi$' for _ in np.arange(0, 3.5, 0.5)]
_ = plt.xticks(xticks, xlabels)
```

## Preparing a bivariate random variable

```python
X1 = rng.normal(size=(1000, 1))/2
X2 = rng.normal(size=(1000, 1))
X = np.concatenate([X1,X2], axis=1)

Y1 = rng.normal(size=(1000, 1))
Y2 = rng.normal(size=(1000, 1))
Y = np.concatenate([Y1,Y2], axis=1)
```

```python

fig, ax = plt.subplots(figsize=(8,8))

divider = make_axes_locatable(ax)
ax_histx = divider.append_axes("top", 1.0, pad=0.1, sharex=ax)
ax_histy = divider.append_axes("right", 1.0, pad=0.1, sharey=ax)

ax_histx.xaxis.set_tick_params(labelbottom=False)
ax_histy.yaxis.set_tick_params(labelleft=False)
ax.scatter(X1, Y1)
ax.scatter(X2, Y2, marker='+')
ax.set_aspect(1.0)

ax_histx.hist(X, bins=20)
ax_histy.hist(Y, bins=20, orientation='horizontal')
# ax_histx.hist(X, bins=20)
# ax_histy.hist(Y1, bins=20, orientation='horizontal')
```
